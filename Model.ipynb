```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("phishing.csv")
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
X= df.drop(columns=['class','Index'])
X.head()
```

```python
X = df[['UsingIP', 'LongURL', 'ShortURL', 'Symbol@', 'Redirecting//', 'PrefixSuffix-', 'SubDomains', 'HTTPS', 'NonStdPort', 'AbnormalURL']]
X.head()
```

```python
Y=df['class']
Y=pd.DataFrame(Y)
Y.head()
```

```python
train_X,test_X,train_Y,test_Y=train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
print(train_X.shape)
print(test_X.shape)
print(train_Y.shape)
print(test_Y.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
rfc=RandomForestClassifier()
model_4=rfc.fit(train_X,train_Y)
```

```python
rfc_predict=model_4.predict(test_X)
```

```python
accuracy_score(rfc_predict,test_Y)
```

```python
print(classification_report(rfc_predict,test_Y))
```

```python
import pickle

# Save the model to a pickle file
with open('phishing_detection_model.pkl', 'wb') as file:
    pickle.dump(model_4, file)
```
